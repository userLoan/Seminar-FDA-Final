## Fundamentals

```python
import pandas as pd
import numpy as np
import yfinance as yf
from time import sleep
from math import isnan


# 2) DataFrame gốc
stocks = pd.DataFrame(stock, columns=["Symbol","Name"]).set_index("Symbol")

# 3) Map sang mã Yahoo (mặc định HOSE là .VN)
stocks["YahooSymbol"] = stocks.index.map(lambda s: f"{s}.VN")

wanted = ["marketCap","trailingPE","forwardPE","pegRatio","priceToBook",
          "trailingEps","forwardEps","bookValue","payoutRatio","beta",
          "fiveYearAvgDividendYield","52WeekChange","averageVolume",
          "enterpriseToRevenue","profitMargins"]

rows = []
for sym, row in stocks["YahooSymbol"].items():
    t = yf.Ticker(row)
    info = {}
    # retry nhỏ cho đỡ lỗi mạng
    for _ in range(2):
        try:
            info = t.info or {}
            break
        except Exception:
            sleep(0.6)

    rec = {"Symbol": sym}
    # lấy thẳng từ info
    for k in wanted:
        rec[k] = info.get(k, np.nan)

    # Fallback 1: 52W change & averageVolume từ lịch sử nếu thiếu
    need_52w = pd.isna(rec["52WeekChange"])
    need_vol = pd.isna(rec["averageVolume"])
    if need_52w or need_vol:
        try:
            h = t.history(period="1y", interval="1d", auto_adjust=False)
            if not h.empty:
                if need_52w:
                    first, last = float(h["Close"].iloc[0]), float(h["Close"].iloc[-1])
                    rec["52WeekChange"] = (last/first - 1.0) if first else np.nan
                if need_vol:
                    rec["averageVolume"] = float(h["Volume"].mean())
        except Exception:
            pass

    # Fallback 2: enterpriseToRevenue nếu thiếu (EV/Revenue)
    if pd.isna(rec["enterpriseToRevenue"]):
        ev  = info.get("enterpriseValue", np.nan)
        rev = info.get("totalRevenue", info.get("revenue", np.nan))
        try:
            if not pd.isna(ev) and not pd.isna(rev) and rev:
                rec["enterpriseToRevenue"] = float(ev)/float(rev)
        except Exception:
            pass

    rows.append(rec)

fundamentals = pd.DataFrame(rows).set_index("Symbol")

# 4) Chỉ giữ các cột thực sự có mặt
available_cols = [c for c in wanted if c in fundamentals.columns]
fundamentals = fundamentals[available_cols].fillna(0)

fundamentals.head()
```

## Historical values

```python
import pandas as pd
import numpy as np
import yfinance as yf
from time import sleep
from math import isnan
import re
from datetime import datetime
from pandas.tseries.offsets import DateOffset


# 2) DataFrame gốc
stocks = pd.DataFrame(stock, columns=["Symbol","Name"]).set_index("Symbol")

# 3) Map sang mã Yahoo (mặc định HOSE là .VN)
stocks["YahooSymbol"] = stocks.index.map(lambda s: f"{s}.VN")

wanted = ["marketCap","trailingPE","forwardPE","pegRatio","priceToBook",
          "trailingEps","forwardEps","bookValue","payoutRatio","beta",
          "fiveYearAvgDividendYield","52WeekChange","averageVolume",
          "enterpriseToRevenue","profitMargins"]

rows = []
for sym, row in stocks["YahooSymbol"].items():
    t = yf.Ticker(row)
    info = {}
    # retry nhỏ cho đỡ lỗi mạng
    for _ in range(2):
        try:
            info = t.info or {}
            break
        except Exception:
            sleep(0.6)

    rec = {"Symbol": sym}
    # lấy thẳng từ info
    for k in wanted:
        rec[k] = info.get(k, np.nan)

    # Fallback 1: 52W change & averageVolume từ lịch sử nếu thiếu
    need_52w = pd.isna(rec["52WeekChange"])
    need_vol = pd.isna(rec["averageVolume"])
    if need_52w or need_vol:
        try:
            h = t.history(period="1y", interval="1d", auto_adjust=False)
            if not h.empty:
                if need_52w:
                    first, last = float(h["Close"].iloc[0]), float(h["Close"].iloc[-1])
                    rec["52WeekChange"] = (last/first - 1.0) if first else np.nan
                if need_vol:
                    rec["averageVolume"] = float(h["Volume"].mean())
        except Exception:
            pass

    # Fallback 2: enterpriseToRevenue nếu thiếu (EV/Revenue)
    if pd.isna(rec["enterpriseToRevenue"]):
        ev  = info.get("enterpriseValue", np.nan)
        rev = info.get("totalRevenue", info.get("revenue", np.nan))
        try:
            if not pd.isna(ev) and not pd.isna(rev) and rev:
                rec["enterpriseToRevenue"] = float(ev)/float(rev)
        except Exception:
            pass

    rows.append(rec)

fundamentals = pd.DataFrame(rows).set_index("Symbol")

# 4) Chỉ giữ các cột thực sự có mặt
available_cols = [c for c in wanted if c in fundamentals.columns]
fundamentals = fundamentals[available_cols].fillna(0)

fundamentals.head()

def clean_sym(s):               # strip $, ^, spaces...
    return re.sub(r'^[\$\^]+', '', s).strip().upper()

def yahoo_sym(s):               # add .VN for HOSE
    return f"{clean_sym(s)}.VN"

yahoo_symbols = [yahoo_sym(s[0]) for s in stock] # changed VN100 to stock and access the ticker symbol

# Use period="max" (ổn định hơn), rồi cắt 5 năm gần nhất
raw = yf.download(
    tickers=" ".join(yahoo_symbols),
    period="max",
    interval="1d",
    auto_adjust=False,
    actions=False,
    threads=True,
    group_by="ticker",
    progress=False
)

# tiện: ngày mốc 5 năm
cutoff = pd.Timestamp(datetime.today().date()) - DateOffset(years=5)

# ==== 4) Helper indicators ====
def rsi(series, window=14):
    delta = series.diff()
    up = delta.clip(lower=0)
    down = -delta.clip(upper=0)
    roll_up = up.ewm(alpha=1/window, adjust=False).mean()
    roll_down = down.ewm(alpha=1/window, adjust=False).mean()
    rs = roll_up / roll_down
    return 100 - (100 / (1 + rs))

def macd(series, fast=12, slow=26, signal=9):
    ema_fast = series.ewm(span=fast, adjust=False).mean()
    ema_slow = series.ewm(span=slow, adjust=False).mean()
    macd_line = ema_fast - ema_slow
    signal_line = macd_line.ewm(span=signal, adjust=False).mean()
    hist = macd_line - signal_line
    return macd_line, signal_line, hist

# ==== 5) Build per-ticker feature DataFrames ====
frames = {}
for sym in stock:
    ys = yahoo_sym(sym[0]) # added indexing to access the ticker symbol
    try:
        df = raw[ys].copy() if (ys in raw.columns.get_level_values(0)) else pd.DataFrame()
    except Exception:
        df = pd.DataFrame()

    if df.empty:
        continue

    # If single-level cols (when only one ticker), normalize names
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.droplevel(0)
    # keep last 5Y if available
    df = df.loc[df.index >= cutoff].dropna(how="all")
    if df.empty or "Close" not in df.columns:
        continue

    # ==== your features ====
    close = df["Close"]
    df["NormClose"] = (close - close.mean()) / close.std(ddof=0)

    # daily log return
    dlog = np.log(1 + close.pct_change())
    df["DailyLogReturn"] = dlog

    df["ALR1W"] = dlog.rolling(window=5).sum() * 5
    df["ALR2W"] = dlog.rolling(window=10).sum() * 5
    df["ALR1M"] = dlog.rolling(window=21).sum() * 21
    df["ALR2M"] = dlog.rolling(window=42).sum() * 21

    # normalize by stdev of DailyLogReturn
    std = dlog.std(ddof=0)
    if std and not np.isnan(std) and std != 0:
        for c in ["DailyLogReturn","ALR1W","ALR2W","ALR1M","ALR2M"]:
            df[c] = df[c] / std

    # RSI / MACD
    df["RSI"] = rsi(close) / 100.0
    macd_line, signal_line, hist = macd(close)
    df["MACD"] = macd_line

    # Drop columns you don’t need
    for c in ["Open","High","Low","Volume"]:
        if c in df.columns:
            df.drop(columns=[c], inplace=True)

    # Add the processed DataFrame to the frames dictionary
    frames[sym[0]] = df # Use the ticker symbol as the key

values = pd.concat(frames, names=["Symbol", "Date"]).dropna(how="all")
print(values.head(100))
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

## Group by Sector

After collecting and pre-processing the data, the graphs are created to build a network between the stocks, using their sectors and correlation.

```python
adj_stocks = np.array([
	[stocks.loc[stock1, 'Sector'] == stocks.loc[stock2, 'Sector'] * (stock1 != stock2) for stock1 in stocks.index] for stock2 in stocks.index
]).astype(int)
stocks_graph = nx.from_numpy_array(adj_stocks)
stocks_graph = nx.relabel_nodes(stocks_graph, dict(enumerate(stocks.index)))

plt.figure(figsize=(12, 5))
nx.draw(stocks_graph, with_labels=True, node_size=500, node_color='skyblue', font_size=8, font_weight='bold', font_color='black', pos=nx.spring_layout(stocks_graph, k=.5))
plt.title('Stocks Graph by Sector')
plt.show()
```

## Graph by fundamentals correlation
A great way to link the stocks would be using their fundamental data, such as market capitalization, p/e ratio, etc. An autoencoder is built to reduce the dimensionality of the fundamental data, and the stocks are clustered and linked following these clusters. Furthermore, the correlation between the fundamental data is used to link the stocks that have a high correlation.

```python
fundamentals = pd.read_csv('/content/fundamentals.csv').set_index("Symbol")
fundamentals.head(n=10)
```

```python
def get_clusters(data: np.ndarray, n_clusters: int) -> np.ndarray:
    kmeans = KMeans(n_clusters=n_clusters)
    return kmeans.fit(data).labels_
fundamentals_corr = fundamentals.transpose().corr(method="spearman")
fundamentals_corr = (fundamentals_corr - (fundamentals_corr == 1))  # Remove self-correlation
fundamentals_corr.head(n=10)
```

```python
plt.figure(figsize=(10, 3.5))
plt.hist(fundamentals_corr.values.flatten(), bins=50, color='skyblue', edgecolor='black')
plt.title('Fundamentals Correlation Distribution')
plt.xlabel('Correlation')
plt.ylabel('Frequency')
plt.show()
```

```python
fundamentals_corr_np = fundamentals_corr.to_numpy()
adj_fundamentals_corr = (fundamentals_corr_np * (abs(fundamentals_corr_np) > .7).astype(int))

fundamentals_corr_graph = nx.from_numpy_array(adj_fundamentals_corr)
fundamentals_corr_graph = nx.relabel_nodes(fundamentals_corr_graph, dict(enumerate(fundamentals_corr.index)))

plt.figure(figsize=(12, 5))
nx.draw(fundamentals_corr_graph, with_labels=True, node_size=500, node_color='skyblue', font_size=8, font_weight='bold', font_color='black', pos=nx.spring_layout(fundamentals_corr_graph))
plt.title('Stocks Graph by Fundamentals Correlation')
plt.show()
```

## Merging the graphs
To make use of the two graphs to build the final adjacency matrix. A correlation bonus is given to two stocks sharing the same sector.

```python
corr_threshold, sector_bonus = .7, .05

share_sector = pd.get_dummies(stocks[["Sector"]]).transpose().corr().to_numpy().astype(int) - np.eye(len(stocks), dtype=int)
# abs because GCNConv only accepts positive weights
adj = abs(fundamentals_corr_np) + share_sector * sector_bonus
adj = adj * (abs(adj) > corr_threshold)
adj = adj / adj.max()

graph = nx.from_numpy_array(adj)
graph = nx.relabel_nodes(graph, dict(enumerate(stocks.index)))

plt.figure(figsize=(12, 5))
nx.draw(graph, with_labels=True, node_size=500, node_color='skyblue', font_size=8, font_weight='bold', font_color='black', pos=nx.spring_layout(graph, k=.5))
plt.title(f'Stocks Graph (Mean degree: {np.mean([degree for node, degree in graph.degree]):.2f}, Density: {nx.density(graph):.4f})')
plt.show()
```

The feature matrix is a 3D matrix of shape (stocks_nb, features_nb, timestamps_nb). The adjacency matrix is used to create the edge_index matrix in the PyTorch Geomtric format.

```python
nodes_nb = len(adj)

# Determine the maximum number of timestamps across all stocks
max_timestamps = values.groupby(level='Symbol').size().max()

# Reshape and pad/truncate each stock's data to have a consistent number of timestamps
x_list = []
for symbol in values.index.get_level_values('Symbol').unique():
    stock_data = values.loc[symbol].drop(columns=["Close"]).to_numpy()
    # Pad or truncate to max_timestamps
    if stock_data.shape[0] < max_timestamps:
        padding = np.full((max_timestamps - stock_data.shape[0], stock_data.shape[1]), np.nan)
        stock_data = np.vstack((stock_data, padding))
    elif stock_data.shape[0] > max_timestamps:
        stock_data = stock_data[:max_timestamps, :]
    x_list.append(stock_data)

x = np.array(x_list) # shape (nodes_nb, timestamps_nb, features_nb)
x = np.swapaxes(x, 1, 2)  # shape (nodes_nb, features_nb, timestamps_nb)

# Fill potential NaNs introduced by padding
x = np.nan_to_num(x, nan=0.0)


edge_nb = np.count_nonzero(adj)
edge_index = np.zeros((2, edge_nb))
edge_weight = np.zeros((edge_nb,))
count = 0
for i in range(nodes_nb):
        for j in range(nodes_nb):
                if (weight := adj[i, j]) != 0:
                        edge_index[0, count], edge_index[1, count] = i, j
                        edge_weight[count] = weight
                        count += 1
x.shape, edge_index.shape, edge_weight.shape
```

## Defining the PyTorch Geometric datasets


```python
import numpy as np
import pandas as pd
import torch


def get_graph_in_pyg_format(values_path: str, adj_path: str) -> tuple[torch.Tensor, torch.Tensor, torch.Tensor, torch.Tensor]:
	"""
	Creates the PyTorch Geometric graph data from the stock price data and adjacency matrix.
	:param values_path: Path of the CSV file containing the stock price data
	:param adj_path: Path of the NumPy file containing the adjacency matrix
	:return: The graph data in PyTorch Geometric format
		x: Node features (nodes_nb, timestamps_nb, features_nb)
		close_prices: Close prices (nodes_nb, timestamps_nb)
		edge_index: Edge index (2, edge_nb)
		edge_weight: Edge weight (edge_nb,)
	"""
	values = pd.read_csv(values_path).set_index(['Symbol', 'Date'])
	adj = np.load(adj_path)
	nodes_nb, edge_nb = len(adj), np.count_nonzero(adj)
	x = torch.tensor(
		values.drop(columns=["Close"]).to_numpy().reshape((nodes_nb, -1, values.shape[1] - 1)), dtype=torch.float32
	)
	x = x.transpose(1, 2)
	close_prices = torch.tensor(
		values[["Close"]].to_numpy().reshape((nodes_nb, -1)), dtype=torch.float32
	)
	edge_index, edge_weight = torch.zeros((2, edge_nb), dtype=torch.long), torch.zeros((edge_nb,), dtype=torch.float32)
	count = 0
	for i in range(nodes_nb):
		for j in range(nodes_nb):
			if (weight := adj[i, j]) != 0:
				edge_index[0, count], edge_index[1, count] = i, j
				edge_weight[count] = weight
				count += 1

	return x, close_prices, edge_index, edge_weight


def get_stocks_labels() -> list[str]:
	"""
	Retrieves the labels (symbols) of the dataset stocks
	:return: The list of stock labels
	"""
	return pd.read_csv('/content/VN100.csv')["Symbol"].unique().tolist()
```

```python
# =========================
# VN100 graph dataset (PyG)
# =========================
import os, os.path as osp
from typing import Callable, Optional, Tuple, List

import numpy as np
import pandas as pd
import torch
from torch_geometric.data import Dataset, Data


# ---------- Utils: CSV(long) -> (X, close, edge_index, edge_weight) ----------
def get_graph_in_pyg_format(
    values_path: str,
    adj_path: str,
    wanted_features: List[str] = None,
) -> Tuple[torch.Tensor, torch.Tensor, torch.Tensor, torch.Tensor, List[str]]:
    """
    Trả về:
        X            : (N, F, T) float32
        close_prices : (N, T)     float32
        edge_index   : (2, E)     int64
        edge_weight  : (E,)       float32
        feature_names: list[str]  độ dài F (thứ tự kênh trong X)
    """
    df = pd.read_csv(values_path)
    # loại bỏ cột thừa kiểu Unnamed
    df = df.loc[:, ~df.columns.str.contains(r"^Unnamed")]
    # ép kiểu thời gian & sort
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values(["Date", "Symbol"])

    # danh sách feature mong muốn (giữ cột nào thật sự tồn tại)
    default_feats = ["NormClose","DailyLogReturn","ALR1W","ALR2W","ALR1M","ALR2M","RSI","MACD"]
    wanted = default_feats if wanted_features is None else wanted_features
    feats = [c for c in wanted if c in df.columns]
    if not feats:  # fallback nếu chưa tạo các chỉ báo
        feats = ["Close"]

    # chọn thứ tự mã (N) theo pivot của DailyLogReturn/Close
    base_col = "DailyLogReturn" if "DailyLogReturn" in df.columns else "Close"
    base = df.pivot(index="Date", columns="Symbol", values=base_col).sort_index()
    symbols = base.columns.tolist()

    # gom ngày chung cho mọi feature (hoặc union nếu giao rỗng)
    mats, date_sets = {}, []
    for f in feats:
        m = df.pivot(index="Date", columns="Symbol", values=f)
        mats[f] = m
        date_sets.append(set(m.index))
    common_dates = sorted(set.intersection(*date_sets)) or sorted(set().union(*[set(m.index) for m in mats.values()]))

    # build tensor (N,F,T)
    stacks = []
    for f in feats:
        m = mats[f].reindex(index=common_dates, columns=symbols).astype(float)
        m = m.ffill().bfill()                   # xử lý thiếu
        stacks.append(m.to_numpy())             # (T,N)
    X_TFN = np.stack(stacks, axis=1)            # (T,F,N)
    X = np.moveaxis(X_TFN, (0,1,2), (2,1,0)).astype(np.float32)  # -> (N,F,T)

    # close prices (N,T)
    close = df.pivot(index="Date", columns="Symbol", values="Close") \
              .reindex(index=common_dates, columns=symbols).ffill().bfill()
    close_prices = close.to_numpy().T.astype(np.float32)

    # edges
    adj = np.load(adj_path)
    A = adj[..., -1] if adj.ndim == 3 else adj
    if A.shape[0] != len(symbols):
        raise ValueError(f"Adj size {A.shape} != number of symbols {len(symbols)}. Hãy tính adj từ đúng pivot.")
    src, dst = np.where(A > 0)
    edge_index = np.vstack([src, dst]).astype(np.int64)
    edge_weight = A[src, dst].astype(np.float32)

    return (torch.from_numpy(X),
            torch.from_numpy(close_prices),
            torch.from_numpy(edge_index),
            torch.from_numpy(edge_weight),
            feats)


# ---------- PyG Dataset ----------
class VN100Stocks(Dataset):
    """
    Graph time-series dataset cho VN100.
    - values_path: CSV 'dài' có cột Symbol, Date, Close, ... (các feature)
    - adj_path   : NPY (N×N) hoặc (N×N×T)
    - target_feature: tên feature trong X dùng làm nhãn y (mặc định 'NormClose'; nếu không có sẽ fallback cột đầu tiên)
    """
    def __init__(
        self,
        root: str = "/content/processed",
        values_path: str = "/content/VN100.csv",
        adj_path: str = "/content/adj.npy",
        past_window: int = 25,
        future_window: int = 1,
        target_feature: str = "NormClose",
        force_reload: bool = False,
        transform: Optional[Callable] = None,
    ):
        self.values_path = values_path
        self.adj_path = adj_path
        self.past_window = past_window
        self.future_window = future_window
        self.target_feature = target_feature

        # cache shapes để tính len() & filenames
        x, close, _, _, feats = get_graph_in_pyg_format(self.values_path, self.adj_path)
        self._N, self._F, self._T = x.shape # Set _T here
        self._feature_names = feats
        self._target_idx = feats.index(target_feature) if target_feature in feats else 0

        super().__init__(root, transform=transform, pre_transform=None,
                         pre_filter=None, force_reload=force_reload)


    @property
    def raw_file_names(self):
        return []  # không dùng raw_dir

    @property
    def processed_file_names(self):
        L = max(self._T - self.past_window - self.future_window, 0)
        return [f"timestep_{i}.pt" for i in range(L)]

    def download(self):
        pass

    def process(self):
        os.makedirs(self.processed_dir, exist_ok=True)
        x, close_prices, edge_index, edge_weight, feats = get_graph_in_pyg_format(
            self.values_path, self.adj_path
        )
        tgt = feats.index(self.target_feature) if self.target_feature in feats else 0

        L = x.shape[2] - self.past_window - self.future_window
        for idx in range(L):
            data = Data(
                x=x[:, :, idx:idx + self.past_window],                       # (N,F,past)
                edge_index=edge_index,                                       # (2,E)
                edge_weight=edge_weight,                                     # (E,)
                close_price=close_prices[:, idx:idx + self.past_window],     # (N,past)
                y=x[:, tgt, idx + self.past_window : idx + self.past_window + self.future_window],  # (N,future)
                close_price_y=close_prices[:, idx + self.past_window : idx + self.past_window + self.future_window],
            )
            torch.save(data, osp.join(self.processed_dir, f"timestep_{idx}.pt"))

    def len(self) -> int:
        return max(self._T - self.past_window - self.future_window, 0)

    def get(self, idx: int) -> Data:
        # PyTorch 2.6+: cần weights_only=False cho object không phải model weights
        return torch.load(
            osp.join(self.processed_dir, f"timestep_{idx}.pt"),
            map_location="cpu",
            weights_only=False,
        )


# ---------- Quick check ----------
if __name__ == "__main__":
    # kiểm tra util
    x, close, ei, ew, feats = get_graph_in_pyg_format('/content/VN100.csv', '/content/adj.npy')
    print("X (N,F,T):", x.shape, "| Close (N,T):", close.shape, "| Edges:", ei.shape, ew.shape, "| Features:", feats)

    # khởi tạo dataset & lấy 1 sample
    ds = VN100Stocks(
        root="/content/processed",
        values_path="/content/VN100.csv",
        adj_path="/content/adj.npy",
        past_window=25,
        future_window=1,
        target_feature="NormClose",  # đổi sang 'Close' nếu muốn dự báo giá
        force_reload=True,           # build lại các timestep_*.pt
    )
    print("samples:", len(ds))
    d0 = ds[0]
    print("x:", d0.x.shape, "edge_index:", d0.edge_index.shape, "edge_weight:", d0.edge_weight.shape,
          "y:", d0.y.shape, "close_price:", d0.close_price.shape)
```
